```python
import pandas as pd

matches = pd.read_csv("../data/WWE_History_1000_Cleaned.csv")
features = pd.read_csv("../data/wrestler_features_2023.csv")

matches = matches.merge(
    features,
    left_on="Winner",
    right_on="wrestlers",
    how="left",
    suffixes=("", "_winner")
)

matches.rename(columns={
    "score": "winner_score",
    "win_rate": "winner_win_rate",
    "title_wins": "winner_title_wins",
    "total_matches": "winner_matches"
}, inplace=True)
```

```python
matches = matches.merge(
    features,
    left_on="Loser",
    right_on="wrestlers",
    how="left",
    suffixes=("", "_loser")
)
```

```python
matches.rename(columns={
    "score": "loser_score",
    "win_rate": "loser_win_rate",
    "title_wins": "loser_title_wins",
    "total_matches": "loser_matches"
}, inplace=True)
```

```python
matches["score_diff"] = matches["winner_score"] - matches["loser_score"]
matches["win_rate_diff"] = matches["winner_win_rate"] - matches["loser_win_rate"]
matches["title_wins_diff"] = matches["winner_title_wins"] - matches["loser_title_wins"]
matches["matches_diff"] = matches["winner_matches"] - matches["loser_matches"]
```

```python
matches["target"] = 1
```

```python
reversed_matches = matches.copy()

reversed_matches["score_diff"] *= -1
reversed_matches["win_rate_diff"] *= -1
reversed_matches["title_wins_diff"] *= -1
reversed_matches["matches_diff"] *= -1
reversed_matches["target"] = 0
```

```python
training_data = pd.concat([matches, reversed_matches], ignore_index=True)
```

```python
training_data[[
    "score_diff",
    "win_rate_diff",
    "title_wins_diff",
    "matches_diff",
    "target"
]].to_csv("../data/match_training_data_2023.csv", index=False)
```
