```python
import pandas as pd
import matplotlib.pyplot as plt

features = pd.read_csv("../data/wrestler_features_2023.csv")
features.head()
features.describe()
```

```python
top_wrestlers = features.sort_values('score', ascending=False)

top10 = top_wrestlers.head(10).reset_index()
top10.rename(columns={'index': 'wrestler'}, inplace=True)

top_wrestlers.head(10)
```

Visualize

```python

plt.figure(figsize=(12,6))
plt.bar(top10['wrestlers'], top10['score'])
plt.title("Top 10 Wrestlers of 2023 by Performance Score")
plt.xlabel("Wrestler")
plt.ylabel("Performance Score")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
