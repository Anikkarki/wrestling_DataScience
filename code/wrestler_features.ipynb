```python
import pandas as pd

df = pd.read_csv("../data/WWE_History_1000_cleaned.csv")
```

```python
wins = df['Winner'].value_counts().rename('wins')
losses = df['Loser'].value_counts().rename('losses')

features = pd.concat([wins, losses], axis=1).fillna(0)

features['win_rate'] = features['wins'] / (features['wins'] + features['losses'])
```

```python
title_wins = (
    df[df['Title Match'] == 1]['Winner']
    .value_counts()
    .rename('title_wins')
)

features = features.merge(
    title_wins,
    left_index=True,
    right_index=True,
    how='left'
)

features['title_wins'] = features['title_wins'].fillna(0)
```

```python
features['total_matches'] = features['wins'] + features['losses']

print(features.columns)
features.head()
```

```python
features['score'] = (
    features['wins'] * 1.0 +
    features['title_wins'] * 2.5 +
    features['win_rate'] * 5
)
features.head()
```

```python
features.to_csv("../data/wrestler_features_2023.csv")
```
